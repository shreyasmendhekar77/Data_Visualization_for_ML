```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_points=np.array([1,2,3,4,5])
y_points=np.array([2,3,5,7,11])
```

```python
#plotting the points
plt.plot(y_points)
```

```python
# marker and labels  - FMT format string
# plt.plot(x_points,y_points,'X--r')  # 'X' is the
plt.plot(x_points,y_points,'X--r')
plt.title("Basic Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# plt.grid(True)
plt.show()
```

```python
# line style and color and size of line 

plt.plot(x_points, y_points, linestyle='--', color='r', linewidth=12)
```

# Subplots - multiple plots


```python



#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,2,1)
plt.plot(x, y, marker='o', linestyle='-', color='b', linewidth=2, markersize=8)

#plot 2:
x1 = np.array([0, 1, 2, 3])
y1 = np.array([10, 20, 30, 40])
plt.subplot(2,2,2)
plt.plot(x1, y1,marker='x',linestyle='--', color='g', linewidth=5)

#plot 3:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([5, 15, 25, 35])
plt.subplot(2,2,3)
plt.plot(x2, y2)

#plot 4:
x3 = np.array([0, 1, 2, 3])
y3 = np.array([2, 4, 6, 8])

plt.subplot(2,2,4)
plt.plot(x3, y3)

```

```python
# Scatter plot 

plt.scatter(x_points, y_points, color='r', marker='o', s=y_points*20 ,alpha=0.5)
```

# Bar plot 

```python
plt.bar(x_points, y_points, color='b', alpha=0.7)
```

# Histogram frequency disribution plots

```python
x_points = np.random.randn(1000)  # Generate 1000 random points
plt.hist(x_points, bins=10, color='g', alpha=1)
```

# Pie Charts

```python
myexplode= [0.1, 0.1, 0.1]  # explode the slices
plt.pie([10, 20, 30], labels=['A', 'B', 'C'], autopct='%1.1f%%', startangle=140,explode=myexplode, shadow=True)
```
